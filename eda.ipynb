## EDA

```python
plt.pie(data['Class'].value_counts(),labels=['Legit','Fraud'],
        autopct='%1.1f%%', startangle=50)
plt.title(f"For total no of transactions : {data['Class'].value_counts().sum()}")
plt.legend()
plt.show()
```

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python

sns.boxplot(y = data.Time,x = data.Class)
plt.show()
```

```python
plt.figure(figsize = (15,10))
plt.subplot(1,2,1)
sns.histplot(legit.Time,kde = True)
plt.xlabel('Time at which transaction was done(in Seconds)')
plt.ylabel('')
plt.title('Legit Transactions')

plt.subplot(1,2,2)
sns.histplot(fraud.Time,kde = True)
plt.title('Fraud Transactions')
plt.xlabel('Time at which transaction was done(in Seconds)')
plt.show()
```

```python
sns.boxplot(y = data.Amount,x = data.Class)
plt.show() 

print("For Actual Transactions\n")
print(legit['Amount'].describe())
print('\n')

print("For Fraud Transactions\n")
print(fraud['Amount'].describe())
print('\n')

```

```python
l_25,l_75 = np.percentile(legit.Amount,[25,75])
f_25,f_75 = np.percentile(fraud.Amount,[25,75])
l_iqr = l_75 - l_25
f_iqr = f_75 - f_25
l_max = np.mean(legit.Amount) + 1.5*l_iqr
f_max = np.mean(fraud.Amount) + 1.5*f_iqr
```

```python
print("Number of outliers in Legit amount transactions: ",
      legit[legit.Amount > l_max].Amount.count())
print("Number of outliers in fraud amount transactions: ",
      fraud[fraud.Amount > f_max].Amount.count())
```

```python
sns.scatterplot(data,y = data.Amount,x = data.Time)
```

```python
l_out = legit[legit.Amount <= l_max]
f_out = fraud[fraud.Amount <= f_max]
```

```python
df_aft_outliers = pd.concat([l_out,f_out])
df_aft_outliers.shape
```
